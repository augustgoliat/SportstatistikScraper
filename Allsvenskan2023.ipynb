```python
def get_team_match_info(list1, list2, matches):
    team_match_info = {}

    # Iterate through list1, list2, and matches simultaneously
    for i, (team, opponent) in enumerate(zip(list1, list2)):
        # Match each team with its opponent and corresponding score
        index = i * 2
        date = matches[index]
        score = matches[index + 1]
        team_match_info[(team, opponent)] = {'date': date, 'score': score}

    return team_match_info

# Example usage:
team_match_info = get_team_match_info(list1, list2, matches)
for teams, info in team_match_info.items():
    # Printing team names, date, and score with Malmö first
    print(f"{teams[1]} - {teams[0]} - Date: {info['date']}, Score: {info['score']}")
```
